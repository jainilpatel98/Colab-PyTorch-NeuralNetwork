```python
import concurrent.futures
from typing import Dict, List
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, PageBreak
from reportlab.lib.styles import getSampleStyleSheet




def create_pdf(report: Dict[str, Dict[str, str]], filename: str = "scouting_report.pdf"):
    """Convert the report to a formatted PDF"""
    doc = SimpleDocTemplate(filename, pagesize=letter)
    styles = getSampleStyleSheet()
    flowables = []
    
    for category, content in report.items():
        # Add category header
        flowables.append(Paragraph(f"<b>{category.upper()}</b>", styles['Heading1']))
        flowables.append(Spacer(1, 12))
        
        for prompt, response in content.items():
            # Add prompt and response
            flowables.append(Paragraph(f"<u>Analysis Request:</u> {prompt}", styles['BodyText']))
            flowables.append(Spacer(1, 6))
            flowables.append(Paragraph(f"<u>Scouting Analysis:</u> {response}", styles['Normal']))
            flowables.append(Spacer(1, 24))
        
        flowables.append(PageBreak())
    
    doc.build(flowables)
```

```python
import concurrent.futures
from typing import Dict, List
from tqdm import tqdm
import dill  # Needed for proper serialization

# Add this at the top with other imports
from functools import partial
import concurrent.futures
import multiprocessing
from typing import Dict, List
from tqdm import tqdm
import dill
import loky  # Better multiprocessing support

# Modified process_category with internal client initialization
def process_category(category: str, prompts: List[str]) -> Dict[str, List[str]]:
    """Process all prompts in a category with local client initialization"""
    results = {}
    
    # Initialize clients inside the process
    gemini_client, qdrant_client, encoder = load_clients()
    
    with tqdm(total=len(prompts), desc=f"{category[:15]:<15}", position=1, leave=False) as pbar:
        with concurrent.futures.ThreadPoolExecutor(max_workers=1) as executor:
            futures = {executor.submit(generate_response, prompt, gemini_client, qdrant_client, encoder): prompt 
                      for prompt in prompts}
            
            for future in concurrent.futures.as_completed(futures):
                prompt = futures[future]
                try:
                    response = future.result()
                    results[prompt] = response
                except Exception as e:
                    results[prompt] = f"Error: {str(e)}"
                finally:
                    pbar.update(1)
    
    return {category: results}

# Modified generate_scouting_report with loky executor
def generate_scouting_report(prompts: Dict[str, List[str]]) -> Dict[str, Dict[str, str]]:
    """Main function with proper multiprocessing setup"""
    full_report = {}
    total_categories = len(prompts)
    
    with tqdm(total=total_categories, desc="Overall Progress", position=0) as main_pbar:
        # Use loky for better serialization
        with loky.get_reusable_executor(
            max_workers=multiprocessing.cpu_count(),
            timeout=3000,
            kill_workers=True
        ) as executor:
            futures = {
                executor.submit(process_category, category, prompts): category
                for category, prompts in prompts.items()
            }
            
            for future in concurrent.futures.as_completed(futures):
                category = futures[future]
                try:
                    category_results = future.result()
                    full_report.update(category_results)
                    main_pbar.update(1)
                    main_pbar.set_postfix_str(f"Completed {category}")
                except Exception as e:
                    print(f"\nFailed category {category}: {str(e)}")
    
    return full_report
# Add this at the top of your file
import multiprocessing
```
